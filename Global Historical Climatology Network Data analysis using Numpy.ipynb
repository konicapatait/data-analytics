Retrieve the Global Historical Climatology Network Daily station data and find the GSN stations and their name 
(Global Climate Observing System(GCOS) Station Network(GSN))

 * Read the Global Historical Climatology Network Daily(GHCND) station file and save it locally.
 * Filter the Global Station Network (GSN) stations.
 * Find the GSN station Id based upon the name

```python
import numpy as np
import matplotlib.pyplot as pp
import seaborn

%matplotlib inline
```

```python
import urllib.request
urllib.request.urlretrieve('ftp://ftp.ncdc.noaa.gov/pub/data/ghcn/daily/ghcnd-stations.txt','stations.txt')
open('stations.txt','r').readlines()[:10]
```

```python
allGSNStations = {}
def filterGSNStations():
    for line in open('stations.txt','r'):
        if 'GSN' in line:
            fields = line.split()
            allGSNStations[fields[0]] = fields[4]
```

```python
filterGSNStations()
allGSNStations
```

```python
def findStationCode(stationName):
    return {code : name for code, name in allGSNStations.items() if stationName in name}
```

```python
findStationCode('NORTH-SALANG')
```
